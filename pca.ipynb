```r
#raw data
df <- read.csv("AirBNB_Taipei_listings_clean_missing.csv", stringsAsFactors = FALSE)
df <- df[, !(colnames(df) == "id")]
head(df)
```

```r
#adjusted data
df_adjust <- as.data.frame(scale(df, center = TRUE, scale = TRUE))
head(df_adjust)
plot(df_adjust$bathrooms, df_adjust$bedrooms)
```

```r
#covarince
cov_mat <- cov(df_adjust)
cov_mat
```

```r
#Eigen values & Eigen vectors
eigen(cov_mat)
```

```r
#PCA
pr.out <- prcomp(df, scale=TRUE, center=TRUE)
summary(pr.out)
```

```r
#the matrix of variable loadings (i.e., a matrix whose columns contain the eigenvectors)
pr.out$rotation
```

```r
biplot(pr.out)
```

```r
#Proportion of Variance Explained
pr.var <- pr.out$sdev^2
pve <- pr.var*100 / sum(pr.var)

plot(pve, xlab = "Principal Component",
     ylab = "Proportion of Variance Explained",
     ylim = c(0, 100), type = "b")
```

```r
plot(cumsum(pve), xlab = "Principal Component",
     ylab = "Cumulative Proportion of Variance Explained",
     ylim = c(0, 100), type = "b")
```
